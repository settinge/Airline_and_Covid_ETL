```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

### Store CSV into DataFrame

```python
airlines = "../ETL-Project/Resources/airlines.csv"
airlines_df = pd.read_csv(airlines)
airlines_not_null = airlines_df.dropna()
airlines_not_null['Airline ID'] = airlines_not_null['Airline ID'].astype(str)
airlines_not_null.head()
```

```python
routes = "../ETL-Project/Resources/routes.csv"
routes_df = pd.read_csv(routes)
routes_not_null = routes_df.loc[(routes_df['Airline ID'] != '/N')]
routes_not_null['Airline ID'] = routes_df['Airline ID'].astype(str)
routes_not_null
```

```python
combined_airline_routes = pd.merge(airlines_not_null, routes_not_null, on='Airline ID')
combined_airline_routes.head()
```

```python
airports = "../ETL-Project/Resources/airports.csv"
airports_df = pd.read_csv(airports)
airports_not_null = airports_df.dropna()
airports_not_null ['Airport ID'] = airports_not_null['Airport ID'].astype(str)
airports_not_null_source = pd.DataFrame({'Source airport ID': airports_not_null['Airport ID'],
                                  'Source Airport Name':airports_not_null['Name'],
                    'Source Country': airports_not_null['Country']})

airports_not_null_destination = pd.DataFrame({'Destination airport ID': airports_not_null['Airport ID'],
                    'Destination Airport Name':airports_not_null['Name'],
                    'Destination Country': airports_not_null['Country']})

airports_not_null_destination
```

```python
combined_airline_routes_airports = pd.merge(combined_airline_routes, airports_not_null_source, on='Source airport ID')

combined_airline_routes_airports_all = pd.merge(combined_airline_routes_airports, airports_not_null_destination, on='Destination airport ID')

combined_airline_routes_airports_all
```

```python
airline_route_data = pd.DataFrame({'Airline_Name':combined_airline_routes_airports_all['Name'],
                                           'Airline_Code':combined_airline_routes_airports_all['Airline'],
                                           'Departure_Airport':combined_airline_routes_airports_all['Source Airport Name'],
                                           'Departure_Airport_Code': combined_airline_routes_airports_all['Source airport'],
                                           'Departure_Country':combined_airline_routes_airports_all['Source Country'],
                                           'Destination_Airport':combined_airline_routes_airports_all['Destination Airport Name'],
                                           'Destination_Airport_Code': combined_airline_routes_airports_all['Destination airport'],
                                           'Destination_Country':combined_airline_routes_airports_all['Destination Country']})

airline_route_data
```

### Connect to local database

```python
rds_connection_string = "postgres:postgres@localhost:5432/airline_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
airline_route_data.to_sql(name='airline_route_data', con=engine, if_exists='append', index=False)
```
